# CCAR Scenario-Based Forecasting

10-30-2025

I provide a mathematical framework for scenario based forecasting, and quantify forecast error from sensitivity misspecification.


---

In the Comprehensive Capital Analysis and Review (CCAR) process, large banks forecast capital and losses under FRB specified regulatory macroeconomic stress scenarios (Baseline, Adverse, Severely Adverse). Banks model outcomes $Y$ (e.g., credit losses or revenues) as a function of macroeconomic drivers and model sensitivities:

$$
Y = X \beta + \epsilon,
$$

where  

- $Y \in \mathbb{R}^{n \times 1}$ is vector of outcomes (e.g., 9 quarterly loss rates)  

- $X \in \mathbb{R}^{n \times p}$ is design matrix of macro variables (e.g., unemployment, GDP, house prices, rates)  

- $\beta \in \mathbb{R}^{p \times 1}$ is a vector of fixed coefficients

- $\epsilon \in \mathbb{R}^{n \times 1}$ is noise

To generate projected outcomes $\tilde{Y}$, CCAR assumes $\tilde{X}$ is exogenous (scenario supplied by the Fed), and estimated sensitivities $\hat{\beta}$ captures the causal response of the institution to those shocks. The scenario-based forecast is thus:

$$
\tilde{Y} = \tilde{X}\hat{\beta}
$$

## 1. Existence and Full Solution Set

Given the model $g(X;\beta)$, I fix $\beta = \hat{\beta}$, set $g(X;\hat{\beta}) = \tilde{Y}$, and solve for all $X$ that produce a desired outcome $\tilde{Y}$:

$$
X \hat{\beta} = \tilde{Y}.
$$

If $\hat{\beta} \neq 0$, each row $x_i^{\top} \in \mathbb{R}^{1 \times p}$ satisfies the scalar constraint:

$$
x_i^{\top} \hat{\beta} = \tilde{y}_i.
$$

Note this fixes only the component of $x_i^{\top}$ along $\hat{\beta}^{\top}$. All directions orthogonal to $\hat{\beta}$ remain free. Thus the full solution set is:

$$
X = \frac{\tilde{Y} \, \hat{\beta}^{\top}}{\hat{\beta}^{\top} \hat{\beta}} + W,
\quad \text{where } W \hat{\beta} = 0.
$$

$W$ can be any $n \times p$ matrix whose columns lie in the subspace orthogonal to $\hat{\beta}$. The minimum-norm solution sets $W = 0_{n \times p}$:

$$
X_{\star} = \frac{\tilde{Y} \, \hat{\beta}^{\top}}{\hat{\beta}^{\top} \hat{\beta}}.
$$

## 2. Implications for CCAR

$Y = X \hat{\beta}$ defines a linear mapping from scenarios to outcomes, where

- $Y$ = projected capital or losses  
- $X$ = macroeconomic and firm-level scenario  
- $\hat{\beta}$ = model sensitivities

Because this mapping is many-to-one, infinitely many $X$ can produce the same $Y$. Scenario paths must be exogenously defined, they cannot be inferred from the model. The nullspace component $W\hat{\beta}=0$ can be interpreted as judgmental overlays and managerial adjustments to the design matrix $X$. Provided $W$ is orthogonal to $\beta$, the adjustment has no effect on forecast $Y$.

CCAR scenario-based forecasting is best thought of as a constrained simulation, not an exercise in causal inference. The process is valid when the macro paths are exogenously defined. Correspondingly, if $\hat{\beta}$ is 'accurate', the forecasts are entirely a function of chosen design matrix $X$, and governance and validation of the model fit $\hat{\beta}$ doesn't depend on the per scenario realized forecasts.

## 3. Sensitivity and Ill-Conditioning

$\hat{\beta}$ is a vector of estimated sensitivities. CCAR forecasts are hence sensitive to misspecification in the model (wrong choice of covariates) and estimation error. This section answers the question of how error in $\beta$'s estimates translates into error in $Y$.

For true vector $\beta$, let $\hat{\beta} = \beta + \Delta \beta$. Thus

$$
Y = X \beta, \qquad \hat{Y} = X \hat{\beta} = X(\beta + \Delta \beta) = X \beta + X \Delta \beta
$$

with forecast error:

$$
\Delta Y = \hat{Y} - Y = X \, \Delta \beta.
$$

To bound $\Delta Y$, I use the matrix 2-norm defined by

$$
\|X\|_2 = \sup_{\|z\|_2 = 1} \|X z\|_2.
$$

By the Singular Value Decomposition,

$$
\|X\|_2 = \sigma_{\max}(X),
$$

where $ \sigma_{\max}(X)$ is the largest singular value of $X$. By the submultiplicative property of the operator norm, which states $\|A B\|_2 \le \|A\|_2 \, \|B\|_2$, it follows that

$$
\|\Delta Y\|_2 = \|X \Delta \beta\|_2
\le \|X\|_2 \, \|\Delta \beta\|_2
= \sigma_{\max}(X) \, \|\Delta \beta\|_2.
$$

Absolute error in $Y$ is bounded as shown above. For relative forecast error, divide by the magnitude of $Y$:

$$
\frac{\|\Delta Y\|_2}{\|Y\|_2}
=
\frac{\|X \Delta \beta\|_2}{\|X \beta\|_2}.
$$

By the previous inequality,

$$
\frac{\|\Delta Y\|_2}{\|Y\|_2}
\le
\frac{\|X\|_2 \, \|\Delta \beta\|_2}{\|X \beta\|_2}
=
\frac{\sigma_{\max}(X) \, \|\beta\|_2}{\|X \beta\|_2}
\cdot
\frac{\|\Delta \beta\|_2}{\|\beta\|_2}.
$$


The local forward condition number

$$
\kappa_{\text{fwd}}(\beta; X)
=
\frac{\sigma_{\max}(X) \, \|\beta\|_2}{\|X \beta\|_2}
$$

measures $Y$'s sensitivity to small relative errors in $\beta$ and is fixed. To set an upper bound, all that remains is $\Delta \beta$ (specifically, how to perturb $\beta$). $\beta$ is a vector of coefficients, so perturbations in some directions naturally are more consequential than others.

Using $X = U \Sigma V^{\top}$, with singular values $\sigma_1 \ge \sigma_2 \ge \cdots \ge \sigma_r > 0$ and right singular vectors $v_i$, for any perturbation $\Delta \beta$:

$$
X \Delta \beta = \sum_{i=1}^{r} \sigma_i \, (v_i^{\top} \Delta \beta) \, u_i.
$$

The change in $Y$ decomposes into orthogonal components scaled by the singular values. The 2-norm is

$$
\|\Delta Y\|_2^2
= \sum_{i=1}^{r} \sigma_i^2 \, (v_i^{\top} \Delta \beta)^2.
$$

Set $\Delta \beta$ such that the perturbation is exactly aligned with the most amplified direction $v_1$ (e.g. $v_i^{\top} \Delta \beta = 0$ for $i\ne 1$), then

$$
\|\Delta Y\|_2 = \sigma_1 \, \|\Delta \beta\|_2,
$$

and therefore

$$
\frac{\|\Delta Y\|_2}{\|Y\|_2}
= \frac{\sigma_1 \, \|\Delta \beta\|_2}{\|X \beta\|_2}
$$

is the worst-case relative forecast error for a perturbation of a given size.

I provide a toy example below, forecasting over 9Q with 4 drivers (unemployment, GDP, HPI, UST 10Y). I align a 5\% perturbation along the most amplified direction $v_1$, and least amplified $v_{\text{min}}$ and compute relative forecast errors for each.

```python
# standard imports
import numpy as np
from numpy.linalg import svd, norm

# toy CCAR design (9Q, 4 drivers + intercept)
n, p = 9, 5

# unemployment path (%)
u = np.array([4.0,4.5,5.5,6.8,7.8,8.5,8.9,9.1,9.0])

# gdp growth (%)
g = np.array([1.2,-2.5,-5.0,-3.0,0.0,1.0,1.5,1.8,2.0])

# housing price index year on year (%)
h = np.array([6.0,3.0,-1.0,-4.0,-6.0,-5.5,-3.0,-1.0,0.5])

# UST 10Y yield (%)
r = np.array([3.5,2.8,2.0,1.8,2.0,2.3,2.7,3.0,3.2])

# construct design matrix
X = np.column_stack([np.ones(n), u, g, h, r])

# set true coefficients, compute Y = XB
beta_true = np.array([0.2, 0.8, -0.4, -0.3, 0.2])
Y = X @ beta_true

# decompose X = USV'
U, s, Vt = svd(X, full_matrices=False)

# capture max and min singular values
sigma_max, sigma_min = s[0], s[-1]

# condition number
kappa2 = sigma_max / sigma_min
V = Vt.T

# 5% perturbation in beta
delta_mag = 0.05 * norm(beta_true)

# most and least amplified directions
v1, vmin = V[:,0], V[:,-1]

# perturb beta
delta_beta_v1 = delta_mag * v1 / norm(v1)
delta_beta_vmin = delta_mag * vmin / norm(vmin)

# compute perturbed forecasts
Yhat_v1 = X @ (beta_true + delta_beta_v1)
Yhat_vmin = X @ (beta_true + delta_beta_vmin)

# compute relative forecast errors
rel_err_v1 = norm(Yhat_v1 - Y) / norm(Y)
rel_err_vmin = norm(Yhat_vmin - Y) / norm(Y)

# check results
print("Singular Values:")
print(f"sigma_max = {np.round(sigma_max,4)}")
print(f"sigma_min = {np.round(sigma_min,4)}")
print(f"\nCondition number:")
print(f"kappa_2(X) = {np.round(kappa2,2)}")
print(f"\nRelative forecast error for 5% β error:")
print(f"\tAlong v1 (high-gain)  = {np.round(rel_err_v1*100,3)}%")
print(f"\tAlong vmin (low-gain) = {np.round(rel_err_vmin*100,3)}%")
```

Note the following:

- The condition number is $\approx$ 480, so the design matrix is extremely ill-conditioned. Highly correlated macroeconomic variables and a modest horizon make the scenario mapping unstable.

- The error along $v_{1}$ shows small misspecification in high-impact sensitivities can drive large errors. A 5% error for unemployment (sensitivity is 0.8) translates to $\approx$ 5.5% forecast deviation.

- In contrast, the error along $v_{\text{min}}$ is 0.01%. When the coefficient direction is orthogonal to the scenario movement, the effect of sensitivity error is negligible.


In a realistic CCAR design, a condition number in the hundreds implies even modest estimation error for sensitivites can change forecasts (e.g. projected losses) by several percent.

## Summary

The model $Y = X \hat{\beta}$ defines a linear projection from macro scenarios to capital outcomes. The inverse space $X(Y) = \frac{Y \hat{\beta}^{\top}}{\hat{\beta}^{\top}\hat{\beta}} + W$ with ($W\hat{\beta}=0$) explains why multiple macro paths can yield the same forecasts, and the forward map’s sensitivity to coefficient errors scales with $\sigma_{\max}(X)$ and the alignment of $\hat{\beta}$ with the row space of $X$. Most importantly, scenario-based CCAR forecasts are a simulation, not an identified structural model. The credibility of those forecasts depends heavily on exogenous scenario paths, not just internal validity.
